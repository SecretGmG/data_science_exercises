```python
weather_df = pd.read_csv("data/owid-co2-data-reduced.csv")

weather_df.rename(columns={"country": "Country", "year": "Year", "co2": "CO2", "temperature_change_from_co2" : "Temperature change from CO2"}, inplace=True)
display(weather_df)
```

```python
# https://seaborn.pydata.org/generated/seaborn.lineplot.html
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill_between.html

sns.set_style('white')

fig, axes = plt.subplots(2,2, figsize=(14, 12))

# temperature change from CO2

axes[0][0].set_title("Temperature change from CO2")
axes[0][0].set_xlim(1900, 2020)

fill_between_kwargs = {'ls' : '-.', 'fc' : 'white', 'ec' : 'gray', 'alpha' : 1, 'label' : 'Standard Error'}
sns.lineplot(ax = axes[0][0], data = weather_df, x = "Year", y = "Temperature change from CO2", errorbar='se', label = 'Average',color = 'red', err_kws=fill_between_kwargs)

# per capita

axes[0][1].set_title("CO2 per capita evolution across the globe")
weather_df['CO2 per capita [tonnes]'] = weather_df['CO2'] / weather_df['population']
sns.lineplot(ax = axes[0][1], data = weather_df, x = "Year", y = "CO2 per capita [tonnes]", 
             hue = 'Country', 
             hue_order=['Asia', 'Europe', 'North America', 'World'], 
             palette=['cyan', 'green', 'purple', 'black']
             )

# The violin plot
axes[1][0].set_title("Share of C02 per source in 2022 across all countries")

violin_df = weather_df[weather_df['Year'] == 2022]
EMISSION_TYPES = ['cement_co2', 'coal_co2', 'gas_co2', 'oil_co2']

totals = violin_df[EMISSION_TYPES].sum()
for source in EMISSION_TYPES:
    violin_df.loc[:,source] = violin_df[source] / totals[source]


#violin_df.loc[:,EMISSION_TYPES] = 100 * violin_df[EMISSION_TYPES] / violin_df[EMISSION_TYPES].sum()

violin_df = violin_df.melt(id_vars='Country', 
                           value_vars=EMISSION_TYPES, 
                           value_name='Percentage of CO2', 
                           var_name='Source of emission')
# Plot the violins (no hue)
sns.violinplot(ax=axes[1][0], 
               data=violin_df, 
               x='Source of emission', 
               y='Percentage of CO2', 
               hue_order = EMISSION_TYPES,
               palette=['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3'],
               inner=None)

weather_df['GDP per capita'] = weather_df['gdp'] / weather_df['population']

axes[1][1].set_title('GDP vs CO2 per capita in 1996 and 2000')
filtered = weather_df[weather_df['Year'].isin([1996, 2000])]
sns.scatterplot(ax = axes[1][1], data = filtered, x = 'GDP per capita', y = 'CO2 per capita [tonnes]', 
                hue = 'Year', 
                hue_order=[1996, 2000], 
                palette=['purple', 'cyan'],
                style='Year',
                style_order=[1996, 2000],
                markers=['o', 'D']
                )
sns.despine()
```
